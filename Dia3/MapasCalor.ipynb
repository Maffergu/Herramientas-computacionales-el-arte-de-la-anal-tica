```python
density = stats.gaussian_kde(data['AveragePrice'])
n, x, _ =plt.hist(data['AveragePrice'], bins=np.linspace(0, 3, 50), 
                   histtype=u'step', density=True)
plt.plot(x, density(x))
plt.show()
```
